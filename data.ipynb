```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta
```

```python
m = 1000
n = 100
```

```python
# Generate m random probability vectors with n entries

np.random.seed(42)
X_random = np.random.rand(m,n)
X_random = np.apply_along_axis(lambda x: x/sum(x), arr=X_random, axis=1)
X_random
```

```python
plt.scatter(np.arange(n), X_random[0,:])
plt.show()
```

```python
# Generate m `continuous' probability vectors with n entries
# Each vector approximates a beta distribution with random parameters

np.random.seed(42)
X_continuous = np.zeros((m,n))

for i in range(m):
    a, b = np.random.exponential(scale=2, size=2) + 1
    x = beta(a,b).pdf(np.arange(n)/n)
    X_continuous[i,:] = x/sum(x)

X_continuous
```

```python
plt.scatter(np.arange(n), X_continuous[0,:])
plt.show()
```
